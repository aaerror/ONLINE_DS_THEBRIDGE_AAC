##### __Ciudades ms pobladas de California__
---

```python
columns = ["city", "population_total"]
most_populated_cities = df_california[columns].sort_values(
		by="population_total", ascending=False)
most_populated_cities[:10]
```

```python
population = most_populated_cities.iloc[:5,1].values/1000
print("Poblacin:", population)

cities = most_populated_cities.iloc[:5,0].values
print("Ciudades:", cities)
```

```python
font_config = {
	"family": "sans-serif",
	"size": "medium",
	"style": "normal",
	"stretch": "normal",
	"weight": "medium"
}

# azul, gris, aceituna, marrn
bar_colors = [
	"blue",
	"gray",
	"red",
	"olive",
	"brown"
]

fig, ax = plt.subplots(figsize=(8,8),
					   frameon=True,
					   dpi=100,
					   linewidth=0)
plot_bar = ax.bar(cities,
				  population,
				  alpha=0.5,
				  align="center",
				  facecolor=bar_colors,
				  width=0.5)
ax.bar_label(plot_bar, padding=5)
ax.set_title("Ciudades ms pobladas de California")
ax.set_xlabel("Ciudades", fontdict=font_config)
ax.set_xticks(cities)
ax.set_xticklabels(labels=cities, fontdict=font_config)
ax.set_ylabel("Habitantes (miles)", fontdict=font_config)
ax.set_ylim(ymin=0, ymax=5000)
ax.set_yticks(ticks=range(0, 6000, 1000))
ax.set_yticklabels(labels=range(0, 6000, 1000), fontdict=font_config)
plt.show()
```

##### __Distribucin de Superficie en $km^2$__

```python
columns = ["city", "area_total_km2"]
cities_area = df_california[columns].sort_values(
		by="area_total_km2", ascending=False)
cities_area.sample(4)
```

```python
font_config = {
	"family": "sans-serif",
	"size": "medium",
	"style": "normal",
	"stretch": "normal",
	"weight": "medium"
}
font_families = ["Bradley Hand ITC", "Noto Sans"]
font_properties = font_manager.FontProperties(family=font_families)

fig, ax = plt.subplots(figsize=(8,8),
					   frameon=True,
					   dpi=100,
					   linewidth=0)
plot_hist = ax.hist(cities_area["area_total_km2"],
					bins=50,
					alpha=0.5,
					facecolor="steelblue")
ax.set_xlabel(r"Superficie ($km^2$)",
			  fontproperties=font_properties,
			  fontsize=18)
ax.set_ylabel("Nmero de ciudades",
			  fontproperties=font_properties,
			  fontsize=18)
ax.legend(["Hist area"])
plt.show()
```

##### __Poblacin por $km^2$ de agua__
---

```python
columns = ["population_total", "area_water_km2"]
population_water_km2 = df_california[columns]
population_water_km2 = population_water_km2.sort_values(
		by="area_water_km2", ascending=False)
```

```python
population_idx, area_idx = population_water_km2.idxmax()
print(f"Max index of:\n\t-population: {population_idx}\n\t-area: {area_idx}")
```

```python
population_water_km2["population_total"] = population_water_km2["population_total"].apply(lambda x: x/1000)
```

```python
population_1, area_1 = population_water_km2.loc[population_idx,:]
population_2, area_2 = population_water_km2.loc[area_idx,:]
```

```python
font_config = {
	"family": "sans-serif",
	"size": "medium",
	"style": "normal",
	"stretch": "normal",
	"weight": "medium"
}
# font_families = ["Bradley Hand ITC", "Noto Sans"]
# font_properties = font_manager.FontProperties(family=font_families)

fig, ax = plt.subplots(figsize=(8,8),
					   frameon=True,
					   dpi=100,
					   linewidth=0)
plot_scatter = ax.scatter(x="area_water_km2",
						  y="population_total",
						  data=population_water_km2,
						  facecolor="green",
						  alpha=0.5)
arrow_properties = {
	"arrowstyle": "->",
	"facecolor": "black",
}
ax.annotate("Outliers",
			xy=(area_1, population_1),
			xytext=(60,3500),
			arrowprops=arrow_properties)

arrow_properties = {
	"arrowstyle": "->",
	"facecolor": "black",
}
ax.annotate("Outliers",
			xy=(area_2, population_2),
			xytext=(60,3500),
			arrowprops=arrow_properties)
ax.set_title("Poblacin por kilometro cuadrado de agua")
ax.set_xlabel(r"rea de agua en la ciudad ($km^2$)")
ax.set_ylabel("Poblacin (miles habitantes)")
plt.show()
```

##### Diagrama de caja de la Altitud de las ciudades de California
---

```python
df_california.columns
```

```python
columns = ["elevation_m"]
elevation = df_california[columns].sort_values(by="elevation_m",
											   ascending=False)
elevation.sample(4)
```

```python
font_config = {
	"family": "sans-serif",
	"size": "medium",
	"style": "normal",
	"stretch": "normal",
	"weight": "medium"
}
# font_families = ["Bradley Hand ITC", "Noto Sans"]
# font_properties = font_manager.FontProperties(family=font_families)

fig, ax = plt.subplots(figsize=(8,8),
					   frameon=True,
					   dpi=100,
					   linewidth=0)
plot_boxplot = ax.boxplot(x="elevation_m",
						  data=elevation,
						  boxprops={"facecolor": "red"},
						  medianprops={"color": "black"},
						  orientation="horizontal",
						  patch_artist=True)
ax.set_title("Diagrama de caja de la Altitud de las ciudades de California")
ax.set_xlabel("Altitud (m)")
ax.set_xticks(range(0, 3000, 500))
ax.set_xlim(xmax=2500)
ax.set_ylabel("")
plt.show()
```
